```python
%matplotlib inline
import os
import numpy as np
import pandas as pd
from alphamind.api import *
from PyFin.api import *
from config import *
from models import *
from matplotlib import rc
from matplotlib import pyplot as plt

plt.style.use("seaborn-poster")
rc('font', **{'family': 'Microsoft Yahei', 'size': 10})
rc('mathtext', **{'default': 'regular'})
rc('legend', **{'frameon': False})
```

```python
start_date = '2000-01-01'
end_date = '2019-06-01'
engine = SqlEngine(alpha_db)

universe_name = 'zz500'
freq = '2b'
freq_1 = '3b'
freq_2 = '5b'
```

```python
universe = Universe(universe_name)
dates = makeSchedule(start_date, end_date, freq, calendar='china.sse')
dx_returns_2 = engine.fetch_dx_return_range(universe, dates=dates, horizon=map_freq(freq))
dx_returns_3 = engine.fetch_dx_return_range(universe, dates=dates, horizon=map_freq(freq_1))
dx_returns_5 = engine.fetch_dx_return_range(universe, dates=dates, horizon=map_freq(freq_2))
```

```python
unicode = lambda code: ('%06d.XSHG' % int(code)) if int(code) < 600000 else ('%06d.XSHE' % int(code))
```

```python
dx_returns_2 = dx_returns_2.rename(columns={"dx": "dx_2"})
dx_returns_2['code'] = dx_returns_2['code'].apply(unicode)
dx_returns_2
```
