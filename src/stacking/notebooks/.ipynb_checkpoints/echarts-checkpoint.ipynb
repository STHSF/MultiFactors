```python
from pyecharts.charts import Bar, Page
from pyecharts import options as opts
from example.commons import Collector, Faker
from pyecharts.charts import Line, Page
import pandas as pd
import numpy as np

C = Collector()
print((Faker.choose()))
print(type(Faker.values()))
print(np.shape(Faker.values()))

@C.funcs
def line_base() -> Line:
    c = (
        Line()
        .add_xaxis(Faker.choose())
        .add_yaxis("商家A", Faker.values())
        .add_yaxis("商家B", Faker.values())
        .set_global_opts(title_opts=opts.TitleOpts(title="Line-基本示例"))
    )
    return c

Page().add(*[fn() for fn, _ in C.charts]).render_notebook()
```

```python
a = pd.DataFrame({'a':[112, 75, 95, 73, 84, 121, 136], 'b':[116, 129, 60, 110, 28, 50, 115]}, index=['衬衫', '毛衣', '领带', '裤子', '风衣', '高跟鞋', '袜子'])
print(list(a.index.values))
print(np.shape(a.a.values))
print(np.shape(a.b.values))


@C.funcs
def line_base() -> Line:
    c = (
        Line()
        .add_xaxis(list(a.index.values))
        .add_yaxis("A", list(a.a.values))
        .add_yaxis("B", list(a.b.values))
        .set_global_opts(title_opts=opts.TitleOpts(title="Line"))
    )
    return c

Page().add(*[fn() for fn, _ in C.charts]).render_notebook()
```
